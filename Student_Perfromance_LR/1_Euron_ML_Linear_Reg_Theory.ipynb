```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df['Extracurricular Activities'] #it needs to be converted as Yes -1, No -0 - done by Label E ncoder
```

```python
le = LabelEncoder() # creating object #it will automatically labels even if it has 10 different classes like yes,no,not sure, not have
df['Extracurricular Activities'] = le.fit_transform(df['Extracurricular Activities'])
```

```python
df.head()
```

```python
#select all x columns indepedenct varaible
x = df[['Hours Studied',	'Previous Scores',	'Extracurricular Activities',	'Sleep Hours'	,'Sample Question Papers Practiced']]
```

```python
y = df['Performance Index'] #dependent variable
```

```python
y
```

```python
#we have to split the data into training data and test data (to evaulate the odel perfromance) #possiible by train_test_split
```

```python
from sklearn.model_selection import train_test_split
```

```python
#here .2 is the percent, 20 % is used for tes remaining 80% used for training
```

```python
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=.2) # it will give four variables
```

```python
x_train
```

```python
#by using the train data we can build model
```

```python
#for better efficency it is better to convert the data into Standard Normal Distribution on guassian curve that will be better so dataset will be noramlized, so STD will be 1, avaeage be 0 /
```

```python
#to standardize the dataset we have to apply zScore
```

```python
#zscore  = (data - avdofdata/std.of data) , this can be done by standscalar lib
```

```python
scaler = StandardScaler()
x_train_scaled =scaler.fit_transform(x_train) #this will make all col data in different format - standarizing the dataset
x_test_scaled =scaler.fit_transform(x_test)
```

```python
#it is better to do satndarizing even without we can do but model learning may not be better
```

```python
#to train the mode
from sklearn.linear_model import LinearRegression
import pickle  #model is a file which needs to be used
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error #evolution metrixes
```

```python
model = LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_predict = model.predict(x_test_scaled)
```

```python
mean_squared_error(y_test, y_predict)
```

```python
r2_score(y_test, y_predict)  #98% accurate
```

```python
with open("student_lr_final_model.pkl","wb") as file:  #store the model in local so we can use it directly, it should also have all pre-processin e have done here so  the test data will alos will udner go the same process
    pickle.dump((model,scaler,le),file)
```
